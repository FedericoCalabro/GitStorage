```python
# Create DataFrame: a 2-dimensional labeled data structure with columns of potentially different types

columns = ['names', 'births']

names = ['Nick', 'Jess', 'Winston', 'Schmidt', 'Cece']
births = [150, 200, 100, 50, 70]
```

```python
dataset = list(zip(names, births))
dataset
```

```python
df = pd.DataFrame(dataset, columns=columns)
df
```

```python
# save dataframe into csv
df.to_csv('dataframe.csv', index = False)
```

```python
# Create a dataset with date, state, birth rate

date = pd.date_range(start='1/1/1980', end='12/31/1981', freq='W-MON')

births = random.randint(0, 1000, len(date))

states = ['NY', 'TX', 'FL', 'GA']
random_states = [states[random.randint(len(states))] for _ in range(len(date))]

dataset = list(zip(random_states, births, date))
```

```python
df = pd.DataFrame(dataset, columns=['States', 'Births', 'Date'])
df.head()
```

```python
df.to_csv('dataset2.csv', index=False)
```

```python
df = pd.read_csv('dataset2.csv', index_col='Date')
df
```
